```python
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
```

```python
# Plot sigmoid function
x = np.arange(-10, 10)
y = 1/(1+np.exp(-x))
plt.plot(x, y)
plt.title('Sigmoid function')
plt.show()
```

```python
# Plot cost function
x = np.arange(0.01, 0.99, 0.001)
y = -np.log(x)
z = -np.log(1-x)

plt.plot(x, y)
plt.xlabel('h')
plt.ylabel('Cost')
plt.title('If y = 1')
plt.show()

plt.plot(x, z)
plt.xlabel('h')
plt.ylabel('Cost')
plt.title('If y = 0')
plt.show()
```
